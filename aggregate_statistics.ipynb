```python
import pandas as pd
df = pd.read_csv("modified_pokemon_data.csv")
df
```

```python
# Groupby a specific data
df.groupby(["Type 1"]).mean().sort_values("HP", ascending=False)
```

```python
# Getting the count by Grouping by specific column values
df["count"] = 1
df.groupby(["Type 1", "Type 2"]).count()["count"]
```

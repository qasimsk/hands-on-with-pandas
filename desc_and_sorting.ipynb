```python
import pandas as pd

df = pd.read_csv("pokemon_data.csv")
df.describe() # Prints all the statistic related data
```

```python
# Sorting
df.sort_values("Name")
# Additional argument can be given for sorting in descending "ascending=False"
```

```python
# Sorting by multiple columns
df.sort_values(["Type 1", "HP"], ascending=[True, False])
# Here we are sorting Type 1 column in ascending and HP in descending 
```

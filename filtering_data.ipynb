```python
import pandas as pd
import re
df = pd.read_csv("modified_pokemon_data.csv")
# Getting data which satisfies multiple conditions
new_df = df.loc[(df["Type 1"] == "Grass") | (df["Type 2"] == "Poison")]
```

```python
new_df = df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"] > 70)]
new_df.to_csv("filtered.csv")
new_df
```

```python
# Reset index for new filtered dataframe
# inplace=True changes the the current dataframe
# drop=True drops the indexes from the old dataframe
new_df.reset_index(drop=True, inplace=True)
new_df.to_csv("filtered.csv")
```

```python
# Check whether contains some data in it
# ~ signs negates it
df.loc[~df["Name"].str.contains("Mega")]
```

```python
# Using regular expression
df.loc[df["Type 1"].str.contains("Fire|Grass", regex=True)]
```

```python
# More of regular expression
df.loc[df["Name"].str.contains("^Pi[a-z]", regex=True)]
```

```python
# Changing the data
# First arg is to get specific column
# Second arg is which column you want to change
df.loc[df["Type 1"] == "Fire", "Type 1"] = "Flamer"
df
```

```python
df.loc[df["Type 1"] == "Flamer", "Type 1"] = "Fire"
df
```

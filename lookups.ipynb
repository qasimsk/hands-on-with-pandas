```python
import pandas as pd

# Reading the csv file
df = pd.read_csv('pokemon_data.csv')
df
```

```python
# Reading the excel file
dfx = pd.read_excel("pokemon_data.xlsx")
dfx
```

```python
# Reading the txt file
# For reading the text file we have to use a delimiter
dft = pd.read_csv("pokemon_data.txt", delimiter="\t")
dft
```

```python
# For reading n lines from starting we use df.head(n)
df.head(10) # Read 10 lines from the start of the dataframe
```

```python
# For reading n lines from the end we use df.tail(n)
df.tail(10) # Read 10 lines from the end of the dataframe
```

```python
# For getting the headers of the dataframe
df.columns
```

```python
# Reading a specific column
df["Name"] # Getting the name column of dataframe
```

```python
# Reading a specific column and getting specific amount of rows from it
df["Name"][0:10] # Here getting from 0th to 10th
```

```python
# Getting multiple columns 
df[["Name", "Type 1", "Attack"]][0:10]
```

```python
# Getting a specific row with given index
df.iloc[4]
```

```python
# Getting specific number of rows from start and end index
df.iloc[4:8]
```

```python
# Getting specific position using iloc
df.iloc[4, 1]
```

```python
# Iterating each row
for index, row in df.iterrows():
  print(index, row["Name"], row["Type 1"])
```

```python
# Getting rows which contains specific data
df.loc[df["Type 1"] == "Fire"] # Here getting the rows where Type 1 of a pokemon is Fire
```
